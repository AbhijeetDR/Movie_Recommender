# Segregating Important columns

# Making proper format for genres, keyword

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
  lst = []
  for i in ast.literal_eval(obj):
    lst.append(i['name'])
  return lst
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

# Making proper format for cast
Lets take main leads only - 3 actors

```python
movies['cast'][0]
```

Inference:

*   cast_id : id given to the actor
*   character: name used in movie
*   gender(1/2) : 1-female, 2-male
*   name : name of the actor


```python
def convert3(obj):
  lst = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter!=3:
      lst.append(i['name'])
      counter+=1
    else:
      break
  return lst
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

# Lets handle Crew now

```python
movies['crew'][0]
```

Inference:
Alot of members working but lets just use the Director for now

```python
def get_director(obj):
  lst = []
  for i in ast.literal_eval(obj):
    if(i['job'] == 'Director'):
      lst.append(i['name'])
      break
  return lst
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

# Removing spaces in all the columns
Like 'Science fiction' = 'ScienceFiction'

```python
def removeSpaces(obj):
  lst = []
  for i in obj:
    lst.append(i.replace(" ", ""))
  return lst
  
```

```python
movies['genres'] = movies['genres'].apply(removeSpaces)
movies['keywords'] = movies['keywords'].apply(removeSpaces)
movies['cast'] = movies['cast'].apply(removeSpaces)
movies['crew'] = movies['crew'].apply(removeSpaces)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords']+movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'title', 'tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

# Text Vectorization


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
  
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman')
```

```python
recommend('Batman Begins')
```
